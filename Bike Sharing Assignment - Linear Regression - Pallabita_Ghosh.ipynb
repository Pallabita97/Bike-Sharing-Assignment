### Creation of Dummy Variable

### Interpretation of coefficients

#### const = the value of constant term is 0.152534 means in the absence of all other predictor variables, bike rentals can be increased by 0.152534 units. 

#### yr = A coefficient value of 0.233748 means one unit increase in year variable increases bike rentals by 0.233748 unit.

#### holiday = A coefficient value of -0.098175 means one unit increase in holiday variable decreases bike rentals by 0.098175 unit.

#### temp = A coefficient value of 0.518784 means one unit increase in temp variable increases bike rentals by 0.518784 unit.

#### windspeed = A coefficient value of -0.160939 means one unit increase in windspeed variable decreases bike rentals by 0.160939 unit.

#### summer = A coefficient value of 0.079796 means one unit increase in summer variable increases bike rentals by 0.079796 unit.

#### winter = A coefficient value of 0.119175 means one unit increase in winter variable increases bike rentals by 0.119175 unit.

#### January = A coefficient value of -0.044985 means one unit increase in January variable decreases bike rentals by 0.044985 unit.

#### September = A coefficient value of 0.098900 means one unit increase in September variable increases bike rentals by 0.098900 unit.

#### Light Rain or Snow = A coefficient value of -0.285487 means one unit increase in Light Rain or Snow variable decreases bike rentals by 0.285487 unit.

#### Misty and Cloudy = A coefficient value of -0.078798 means one unit increase in Misty and Cloudy variable decreases bike rentals by 0.078798 unit.
